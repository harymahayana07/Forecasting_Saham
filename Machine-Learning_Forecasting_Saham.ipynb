### TUGAS AKHIR FORECASTING HARGA SAHAM

#### Import Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Import Data

```python
df = pd.read_csv('harga_saham.csv', index_col="Date", parse_dates=["Date"])
df
```

#### Visualisasi Data Harga Saham Dengan Plot

```python
df.harga_saham.plot(figsize=(15,5))
```

```python
df=df.resample('M').mean()
df
```

```python
df.harga_saham.plot(figsize=(15,5))
```

#### Data Preprocessing

```python
cols = list()
```

```python
for i in range(1, 0, -1):
    cols.append(df.shift(i))
cols
```

```python
for i in range(0, 1):
    cols.append(df.shift(-i))
cols
```

```python
agg = pd.concat(cols, axis=1)
agg
```

```python
agg.dropna(inplace=True)
```

```python
agg
```

#### Train Test Split

```python
train, test=agg.iloc[:-10, :],agg.iloc[-10:, :]
```

```python
train
```

```python
test
```

#### TrainX dan Trainy 

```python
train=train.reset_index()
```

```python
train
```

```python
trainX, trainy = train.iloc[:, :-1], train.iloc[:, -1]
```

```python
trainy
```

```python
trainX=trainX.drop(columns='Date')
```

```python
trainX
```

#### Modeling Machine Learning

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
# model = RandomForestRegressor(n_estimators=1000)
# model = SVR()
# model = KNeighborsRegressor(n_neighbors=4)
model = LinearRegression()
# model = DecisionTreeRegressor(random_state=0)

model.fit(trainX, trainy)
```

#### Prediction / Forecasting

```python
from sklearn.metrics import mean_absolute_error
predictions=[]
actual=[]
for i in range(len(test)):
    testX, testy = test.iloc[i, :-1], test.iloc[i, -1]
    pred = model.predict([testX])
    print("hasil prediksi:"+str(pred[0]))
    predictions.append(pred[0])
    actual.append(testy)
    print("expected= "+str(testy)+ " predicted= "+str(pred))
```

```python
predictions
```

#### Metrics Evaluasi Model

```python
error = mean_absolute_error(actual, predictions)
```

```python
print(error)
```

#### Visualisasi Plot For Actual & Predict

```python
plt.plot(actual, label='Actual')
plt.plot(predictions, label='Predicted')
plt.legend()
plt.show()
```

#### Dataframe Baru

```python
df_visual = pd.DataFrame({
    'date':test.reset_index()['Date'].to_list(),
    'actuals':actual
})
```

```python
df_visual
```

```python
df_visual['pred']=predictions
```

```python
df_visual.set_index('date', inplace=True)
```

```python
df_visual
```

```python
plt.figure(figsize=(15, 5))
plt.plot(df_visual.actuals, "b-",label='actual')
plt.plot(df_visual.pred, "r-",label='prediction')
```

#### Predicted 

```python
result = pd.concat([df, df_visual], axis=1).drop(columns='actuals')
```
